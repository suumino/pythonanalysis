```python
from selenium import webdriver
import pandas as pd
from bs4 import BeautifulSoup
```

```python
browser = webdriver.Chrome("C:chromedriver.exe")

url = "https://www.starbucks.co.kr/store/store_map.do"

browser.get(url)
```

```python
find_btn = "#container > div > form > fieldset > div > section > article.find_store_cont > article > header.loca_search > h3 > a"
browser.find_element_by_css_selector(find_btn).click()
```

```python
jeju_btn = "#container > div > form > fieldset > div > section > article.find_store_cont > article > article:nth-child(4) > div.loca_step1 > div.loca_step1_cont > ul > li:nth-child(16) > a"
browser.find_element_by_css_selector(jeju_btn).click()
```

```python
all_btn = "#mCSB_2_container > ul > li:nth-child(1) > a"
browser.find_element_by_css_selector(all_btn).click()
```

```python
html = browser.page_source
print(html)
```

```python
soup = BeautifulSoup(html, 'html.parser')
print(soup)
```

```python
starbucks_soup_list=soup.select('li.quickResultLstCon')
print(starbucks_soup_list)
```

```python
print(len(starbucks_soup_list))
```

```python
starbucks_store = starbucks_soup_list[0]
```

```python
name = starbucks_store.select('strong')[0].text.strip()
lat = starbucks_store['data-lat']    # 위도
lng = starbucks_store['data-long']   # 경도
# add = starbucks_store.select('p')[0].text.strip()[:-9:1]
# num = starbucks_store.select('p')[0].text.strip()[-9::1]

add = str(starbucks_store.select('p')[0]).split('<br/>')[0].split('>')[1]
num = str(starbucks_store.select('p')[0]).split('<br/>')[1].split('<')[0]
store = starbucks_store.select('i')[0]['class'][0][4:]
print(add)
print(num)
print(store)
```

```python
starbucks_list = []

for item in starbucks_soup_list:
    name = item.select('strong')[0].text.strip()
    lat = item['data-lat']    # 위도
    lng = item['data-long']   # 경도
    store = item.select('i')[0]['class'][0][4:]
    add = item.select('p')[0].text.strip()[:-9:1]
    num = item.select('p')[0].text.strip()[-9::1]
    starbucks_list.append([name, lat, lng, store, num, add])
```

```python
print(starbucks_list)
```

```python
# 데이터 프레임으로 변환

columns = ['매장명', '위도', '경도', '매장 타입', '전화번호', '주소']
jeju_starbucks_df = pd.DataFrame(starbucks_list,columns = columns)
```
